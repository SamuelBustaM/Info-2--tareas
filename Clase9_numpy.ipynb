```python
import numpy as np

v4 = np.zeros(10, dtype=int)
v4[::2] = 2    # índices pares: 0, 2, 4, 6, 8
v4[1::2] = 1   # índices impares: 1, 3, 5, 7, 9
v4
```

```python
import numpy as np

tablero = np.zeros((8, 8), dtype=int)
tablero[1::2, ::2] = 1   # filas impares, columnas pares
tablero[::2, 1::2] = 1   # filas pares, columnas impares
tablero
```

```python
import numpy as np

rng = np.random.default_rng(42)
m6 = np.round(rng.random((5, 5)), 4)
print("Matriz:\n", m6)
print("Mínimo:", np.min(m6))
print("Máximo:", np.max(m6))
```

```python
import numpy as np

m7 = (m6 - np.min(m6)) / (np.max(m6) - np.min(m6))
print("Matriz normalizada:\n", np.round(m7, 4))
print("Nuevo mínimo:", np.min(m7))
print("Nuevo máximo:", np.max(m7))
```
